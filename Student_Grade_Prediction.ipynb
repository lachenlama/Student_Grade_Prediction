```python
fig,((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(16,12))
ax1.bar(df['internet'], df['G3'], color='Orange')
ax1.set_xlabel("Internet availability")
ax1.set_ylabel("Grade")

ax2.bar(df['romantic'], df['G3'])
ax2.set_xlabel("In a relationship")
ax2.set_ylabel("Grade")

ax3.bar(df['nursery'], df['G3'], color='Magenta')
ax3.set_xlabel("Attended Nursery")
ax3.set_ylabel("Grade")

ax4.bar(df['higher'], df['G3'], color='Green')
ax4.set_xlabel("Opts for higher education")
ax4.set_ylabel("Grade")
```

From above we can observe, <br>

In Graph (2,2) students who are in relationship scored less marks. <br>
In Graph (1,1) stuents who attended nursery scored higher. <br>
In Graph (2,1) students who want higher education scored higher.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
features = ['Medu', 'Fedu', 'studytime', 'Dalc', 'G1', 'G2', 'failures', 'absences']
y = df['G3']
X = df[features]
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
lr = LinearRegression()
lr_model = lr.fit(train_X, train_y)
lr_model_pred = lr.predict(val_X)
```

```python
print("Mean Absolute Error: ", mean_absolute_error(val_y, lr_model_pred))
```

```python
print("Accuracy: ", lr.score(val_X, val_y))
```
